```python
circuits_df = (spark.read
                    .format("csv")
                    .option("header", "true")
                    .option("inferSchema","true")
                    .option("mode","FAILFAST")
                    #.schema(circuits_schema)
                    .load(f"{landing_folder_path}/circuits.csv")               
)
```

```python
## ADDING METADATA COLUMNS
circuits_final = add_metadata_columns(circuits_df)
```

```python
## WRITITNG THE DATA TO DELTA TABLE

circuits_final.write\
           .format("delta")\
           .mode("overwrite")\
           .option("overwriteSchema", "true")\
           .saveAsTable(f"{catalog}.{bronze_schema}.circuits")
```
