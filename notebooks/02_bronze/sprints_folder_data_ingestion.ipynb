```python
source_table = f"{landing_folder_path}/sprints"
target_table = f"{catalog}.{bronze_schema}.sprints"
```

```python
sprints_df = spark.read.format("json")\
                  .option("multiLine","true")\
                  .option("inferSchema","true")\
                  .option("mode","FAILFAST")\
                  .load(source_table)
```

```python
sprints_final = add_metadata_columns(sprints_df)
```

```python
sprints_final.write.format("delta")\
             .mode("overwrite")\
             .saveAsTable(target_table)
```
