```python
source_path = f"{landing_folder_path}/constructors.json"
target_table = f"{catalog}.{bronze_schema}.constructors"
```

```python
constructors_df = spark.read.format("json")\
                      .option("inferSchema","true")\
                      .option("mode","FAILFAST")\
                      .load(source_path)
```

```python
constructors_final = add_metadata_columns(constructors_df)
```

```python
constructors_final.write.format("delta")\
                 .mode("overwrite")\
                 .saveAsTable(target_table)
```
