```python
source_path = f"{landing_folder_path}/results"
target_table = f"{catalog}.{bronze_schema}.results"
```

```python
results_df = spark.read.format("json")\
                  .option("inferSchema","true")\
                  .option("mode","FAILFAST")\
                  .load(source_path)
```

```python
results_final = add_metadata_columns(results_df)
```

```python
results_final.write.format("delta")\
             .mode("overwrite")\
             .saveAsTable(target_table)
```
